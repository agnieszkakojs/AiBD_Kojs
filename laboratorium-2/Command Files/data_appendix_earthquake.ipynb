```python
#Dla każdej zmiennej 
def draw_countplot(name):
    ax = sns.countplot(x=name, data=data, order=data[name].value_counts().index)
    ax.set_title(label=name)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=-50)
    plt.tight_layout()
    img_name = name[:3] + '_image'
    plt.savefig(rf'C:\Users\lenovo\Desktop\studia\Semestr V\Anliza i Bazy Danych\AiBD_Kojs\laboratorium-2\Analysis Data\{img_name}.png')
    plt.show()
    

draw_countplot("Worried about earthquakes")
draw_countplot("Worried about the Big One")
draw_countplot('The "Big One" will occur in your lifetime?')
draw_countplot("Experienced an earthquake")
draw_countplot("Take precautions for an earthquake")
draw_countplot("Familar with the San Andreas Fault line")
draw_countplot("Familar with the Yellowstone Supervolcano")
draw_countplot("Age")
draw_countplot("Gender")
draw_countplot("Total household combined money")
draw_countplot("US Region")
```
