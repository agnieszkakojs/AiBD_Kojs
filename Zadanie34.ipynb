```python
import matplotlib.pyplot as plt
import numpy as np

def fun(x):
    y = x**2+5
    return y
x1 = np.linspace(-1,1,200)
x2 = np.linspace(-6,6,200)
x3 = np.linspace(0,5,200)

plt.plot(x1, fun(x1))
plt.title(label= "Wykres na przedziale (-1,1)")
plt.xlabel("O x")
plt.ylabel("O y")
plt.legend("wykres")
plt.show()

plt.plot(x2, fun(x2))
plt.title(label= "Wykres na przedziale (-6,6)")
plt.xlabel("O x")
plt.ylabel("O y")
plt.legend("wykres")
plt.show()

plt.plot(x3, fun(x3))
plt.title(label= "Wykres na przedziale (0,5)")
plt.xlabel("O x")
plt.ylabel("O y")
plt.legend("wykres")
plt.show()
```
