Zdefiniuj poniższą funkcję i sporządź jej wykres dla argumentów z danego przedziału:

$f(x)=x^2+5$

1. $x>-1$  oraz  $x<1$
2. $x>-6$  oraz  $x<6$
3. $x>0$  oraz  $x<5$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Deklaracja fukncji, która zawiera funkcje kwadratową 
def fun(x):
    y = x**2+5
    return y

# Trzy przedziały w jakich mają być sporządzone wykresy 
x1 = np.linspace(-1,1,200)
x2 = np.linspace(-6,6,200)
x3 = np.linspace(0,5,200)

# Tworzenie trzech wykresów za pomocą matplotlib
plt.plot(x1, fun(x1))
plt.title(label= "Wykres funkcji $y = x^2+5$ na przedziale (-1,1)")
plt.xlabel("Argumenty x")
plt.ylabel("Wartość y")
plt.legend("y")
plt.grid()
plt.show()

plt.plot(x2, fun(x2))
plt.title(label= "Wykres funkcji $y = x^2+5$ na przedziale (-6,6)")
plt.xlabel("Argumenty x")
plt.ylabel("Wartość y")
plt.legend("y")
plt.grid()
plt.show()

plt.plot(x3, fun(x3))
plt.title(label= "Wykres funkcji $y = x^2+5$ na przedziale (0,5)")
plt.xlabel("Argumenty x")
plt.ylabel("Wartość y")
plt.legend("y")
plt.grid()
plt.show()
```
